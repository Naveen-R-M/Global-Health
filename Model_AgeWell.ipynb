## Overview

The **AgeWell Respiratory Risk Assessment Framework** is an integrated approach to evaluating the respiratory health of individuals by combining data from:

- **Wearable Devices** – Continuous monitoring of vital physiological parameters.
- **Environmental Sensors & Behavioral Tracking** – Real-time data on air quality, allergens, physical activity, and location.
- **Electronic Health Records (EHR)** – Historical clinical data including pre-existing conditions, medication usage, and recent clinical events.

This multi-modal approach provides a dynamic, actionable risk score that supports proactive intervention and personalized health management.

## Risk Assessment Framework

### 1. Passive Physiological Data (Wearables)
These data points continuously monitor critical physiological health indicators.

#### A. Blood Oxygen Saturation (SpO₂)
| SpO₂ Range | Risk Score | Clinical Significance |
|------------|------------|-----------------------|
| ≥95%       | 0          | Normal oxygenation    |
| 92–94%     | 5          | Mild Hypoxemia (early warning) |
| <92%       | 10         | Significant Hypoxemia (high risk) |

#### B. Respiratory Rate (RR)
| Respiratory Rate         | Risk Score | Clinical Significance |
|--------------------------|------------|-----------------------|
| 12–20 breaths/min        | 0          | Normal                |
| 21–24 or 8–11 breaths/min| 5          | Slight abnormality (needs monitoring) |
| >24 or <8 breaths/min    | 8          | Significant abnormality (possible distress) |

#### C. Heart Rate Variability (HRV)
| HRV Level                       | Risk Score | Clinical Significance                         |
|---------------------------------|------------|-----------------------------------------------|
| High/Normal Variability         | 0          | Healthy autonomic nervous system function     |
| Low variability (chronically low)| 5         | Chronic stress, possible underlying issues    |

#### D. Heart Rate (Resting)
| Resting HR (bpm)                | Risk Score | Clinical Significance |
|---------------------------------|------------|-----------------------|
| 60–100 bpm                      | 0          | Normal resting HR     |
| <60 bpm (if symptomatic)        | 5          | Possible bradycardia (monitor closely) |
| >100 bpm (resting, sustained)   | 7          | Tachycardia (possible respiratory distress) |

#### E. Sleep Quality
| Sleep Indicator                      | Risk Score | Clinical Significance              |
|--------------------------------------|------------|------------------------------------|
| No disturbances, 7–9 hours           | 0          | Optimal                          |
| Occasional interruptions             | 3          | Mild disturbance                 |
| Frequent awakenings, <6 hrs sleep     | 6          | Possible apnea, respiratory dysfunction |

---

### 2. Passive Environmental & Behavioral Data

Collected passively from environmental sensors and behavioral tracking.

#### A. Air Quality Exposure (AQI - PM2.5, NO₂)
| Air Quality Index (AQI)    | Risk Score | Clinical Significance            |
|----------------------------|------------|----------------------------------|
| Good (0–50 AQI)            | 0          | Safe                             |
| Moderate (51–100 AQI)      | 4          | Mild irritation, caution advised |
| Unhealthy (>100 AQI)       | 8          | Significant respiratory risk     |

#### B. Allergen Exposure (Pollen, Mold)
| Allergen Level | Risk Score | Clinical Significance                      |
|----------------|------------|--------------------------------------------|
| Low            | 0          | No major risk                              |
| Moderate       | 4          | Potential symptom triggers                 |
| High           | 8          | High risk (asthma/COPD exacerbation risk)    |

#### C. Physical Activity Level (via Accelerometer)
| Activity Level per Week                     | Risk Score | Clinical Significance                      |
|---------------------------------------------|------------|--------------------------------------------|
| Active (>150 mins moderate intensity)       | 0          | Healthy respiratory status                 |
| Moderate (75–149 mins moderate intensity)   | 4          | Mild risk due to reduced fitness           |
| Sedentary (<75 mins moderate intensity)     | 7          | Significant risk, deconditioning possible  |

#### D. Behavioral Patterns (GPS/location data)
| Environment Exposure Frequency                 | Risk Score | Clinical Significance                          |
|------------------------------------------------|------------|------------------------------------------------|
| Frequent access to low-pollution environments  | 0          | Protective                                     |
| Regular visits to moderate-pollution locations | 4          | Increased risk                                 |
| Frequent exposure to high-pollution locations  | 8          | Significant risk                               |

---

### 3. Electronic Health Record (EHR) Data

Integrating EHR data enhances context around existing health conditions, medication usage, and historical health events.

#### A. Pre-existing Conditions
| Condition                                                             | Risk Score | Clinical Significance                  |
|-----------------------------------------------------------------------|------------|----------------------------------------|
| Asthma, mild COPD, bronchitis                                         | 6          | Moderate baseline risk                 |
| Severe COPD, Pulmonary Fibrosis, Chronic Respiratory Disease          | 10         | High baseline risk                     |
| Cardiovascular conditions (CHF, CAD)                                  | 5          | Cardiopulmonary risk interactions      |

#### B. Recent Clinical Events (last 12 months)
| Clinical Event                                 | Risk Score | Clinical Significance                   |
|------------------------------------------------|------------|-----------------------------------------|
| Recent hospitalization (respiratory-related)   | 10         | High risk of readmission                |
| ER visit (respiratory issues)                  | 8          | Elevated immediate risk                 |
| Stable outpatient management                   | 4          | Ongoing risk (lower than hospitalization)|

#### C. Medication Usage
| Medication Type                                       | Risk Score | Clinical Significance                              |
|-------------------------------------------------------|------------|----------------------------------------------------|
| Routine bronchodilator, inhaled steroids              | 5          | Ongoing respiratory management needed            |
| Oxygen therapy                                        | 10         | Significant respiratory impairment               |
| Antibiotics (recent respiratory infection)          | 6          | Increased risk of recurrence or complications      |

## Comprehensive Risk Scoring System

Aggregate scores from the above domains to determine a user’s overall risk:

**Total Respiratory Risk Score** = Physiological Score + Environmental Score + Behavioral Score + EHR Score

| Aggregate Score | Risk Level      | Recommended Actions                                       |
|-----------------|-----------------|-----------------------------------------------------------|
| **0–15**        | **Low Risk**    | Routine monitoring; preventive health education         |
| **16–30**       | **Moderate Risk**| Lifestyle interventions; monthly personalized check-ins   |
| **31–50+**      | **High Risk**   | Immediate clinical review; proactive specialist involvement|

# Task
The framework calculates an overall respiratory risk score by combining three key domains: physiological, environmental/behavioral, and clinical data. First, the physiological score is computed using wearable measurements. For instance, blood oxygen saturation (SpO₂) is scored as 0 for 95% or above, 5 for 92–94%, and 10 for below 92%; similar scoring applies for respiratory rate, heart rate variability, resting heart rate, and sleep quality. Next, the environmental and behavioral score is derived from external factors such as air quality (AQI), allergen exposure, physical activity levels, and location patterns, each contributing points based on risk levels. Lastly, the EHR score incorporates clinical history by scoring pre-existing conditions (e.g., asthma, COPD), recent clinical events (hospitalizations, ER visits), and medication usage. These three scores are added to form the Total Respiratory Risk Score. Depending on the final score, individuals are categorized as low (0–15), moderate (16–30), or high risk (31+), guiding appropriate preventive or clinical interventions.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Update the file paths if needed
personal_health_file = '/content/personal_health_data.csv'
activity_env_file = '/content/activity_environment_data.csv'

# Load the datasets
df_personal = pd.read_csv(personal_health_file)
df_activity = pd.read_csv(activity_env_file)

# Assuming both datasets share the column "User_ID"
df_merged = pd.merge(df_personal, df_activity, on='User_ID', how='inner')

# Create BMI from Weight (kg) and Height (cm)
df_merged['BMI'] = df_merged['Weight'] / ((df_merged['Height'] / 100) ** 2)

# (Optional) Compute Calories_Burned and Distance_Covered if not provided:
if 'Calories_Burned' not in df_merged.columns:
    df_merged['Calories_Burned'] = df_merged['Steps'] * 0.05

if 'Distance_Covered' not in df_merged.columns:
    df_merged['Distance_Covered'] = df_merged['Steps'] * 0.0008

# Create an interaction feature: Blood Oxygen Level with Ambient Temperature
df_merged['Oxy_Ambient_Interaction'] = df_merged['Blood_Oxygen_Level'] * df_merged['Ambient_Temperature']

# Define a risk labeling function that safely handles non-string inputs and NaN values.
def label_risk(row):
    risk_score = 0

    # Blood Oxygen Level
    if row['Blood_Oxygen_Level'] < 93:
        risk_score += 10

    # Medical Conditions: treat missing as "none"
    med_cond = row['Medical_Conditions']
    if pd.isnull(med_cond):
        med_cond = "none"
    else:
        med_cond = str(med_cond).strip().lower()
    if med_cond != "none":
        risk_score += 5

    # Smoker: treat missing as "no"
    smoker = row['Smoker']
    if pd.isnull(smoker):
        smoker = "no"
    else:
        smoker = str(smoker).strip().lower()
    if smoker == "yes":
        risk_score += 4

    # Sleep Duration: low sleep adds risk
    if row['Sleep_Duration'] < 6:
        risk_score += 3

    # Elevated Heart Rate adds risk
    if row['Heart_Rate'] > 100:
        risk_score += 3

    # Simple thresholding for risk level
    if risk_score >= 20:
        return 2  # High risk
    elif risk_score >= 10:
        return 1  # Moderate risk
    else:
        return 0  # Low risk

# If 'risk_level' column does not exist, create it.
if 'risk_level' not in df_merged.columns:
    df_merged['risk_level'] = df_merged.apply(label_risk, axis=1)

# Select features based on your data dictionary.
selected_features = [
    'Age', 'BMI', 'Gender', 'Medical_Conditions', 'Medication', 'Smoker',
    'Alcohol_Consumption', 'Sleep_Duration', 'Deep_Sleep_Duration', 'REM_Sleep_Duration',
    'Wakeups', 'Snoring', 'Heart_Rate', 'Blood_Oxygen_Level', 'ECG', 'Steps',
    'Calories_Burned', 'Distance_Covered', 'Exercise_Type', 'Exercise_Duration',
    'Exercise_Intensity', 'Calories_Intake', 'Water_Intake', 'Stress_Level', 'Mood',
    'Skin_Temperature', 'Ambient_Temperature', 'Battery_Level', 'Body_Fat_Percentage',
    'Muscle_Mass', 'Altitude', 'UV_Exposure',  'Anomaly_Flag', 'Oxy_Ambient_Interaction'
]

# Create a modeling DataFrame.
model_df = df_merged[selected_features + ['risk_level']].copy()

# One-hot encode categorical variables.
categorical_fields = ['Gender', 'Medical_Conditions', 'Medication', 'Smoker',
                      'Alcohol_Consumption', 'Snoring', 'ECG', 'Exercise_Type',
                      'Exercise_Intensity', 'Stress_Level', 'Mood']
model_df = pd.get_dummies(model_df, columns=categorical_fields, drop_first=True)

# Separate features (X) and target (y).
X = model_df.drop(columns='risk_level')
y = model_df['risk_level']

# Standardize numerical features.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train a Random Forest Classifier.
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Evaluate the model.
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Plot a confusion matrix.
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Low', 'Moderate', 'High'],
            yticklabels=['Low', 'Moderate', 'High'])
plt.xlabel("Predicted Risk Level")
plt.ylabel("Actual Risk Level")
plt.title("Confusion Matrix - Risk Levels")
plt.show()


importances = model.feature_importances_
feature_names = X.columns
fi_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)
print("\nFeature Importances:")
print(fi_df)
```
